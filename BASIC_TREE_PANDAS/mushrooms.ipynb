```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import GridSearchCV 
from sklearn.ensemble import RandomForestClassifier
```

```python
mush_df = pd.read_csv('training_mush.csv')
```

```python
mush_df.head()
```

```python
X = mush_df.iloc[:,0:-1]
y = mush_df.iloc[:,-1]
```

```python

clf = RandomForestClassifier(random_state=0)
parametrs = {'n_estimators': [i for i in range(10,50,10)], 'max_depth' : [i for i in range(1,12,2)],\
            'min_samples_leaf':[i for i in range(1,7)], 'min_samples_split':[i for i in range(2,9,2)]}
clf_best = GridSearchCV(clf, parametrs, n_jobs=-1, cv =3)
clf_best.fit(X,y)
```

```python
clf_best.best_params_
```

```python
clf_best.best_score_
```

```python
clf_best_forest = clf_best.best_estimator_
clf_best_forest.feature_importances_

pd.DataFrame({"property":list(mush_df.iloc[:,0:-1]), "value" : clf_best_forest.feature_importances_}).\
sort_values(by = 'value',ascending=False)
```

```python
testing_df = pd.read_csv('testing_mush.csv')
y_pred = clf_best_forest.predict(testing_df)
```

```python
y_correct = pd.read_csv('testing_y_mush.csv')
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_correct, y_pred)
```
